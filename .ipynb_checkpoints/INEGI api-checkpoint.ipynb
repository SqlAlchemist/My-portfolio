```python
import locale
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
import time
```

```python
# path to credentials folder
filename = '../APIS/credentials.json'

# read json file
with open(filename) as f:
    keys = json.load(f)

# read credentials
token = keys['inegi_api']
```

```python
inegi = f'https://www.inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml/INDICATOR/1002000026/es/0700/false/BISE/2.0/{inegi_id}?type=json'
```

```python
inegi
```

```python
pop = '1002000026'
query = 'false'

url = (
        'https://www.inegi.org.mx/app/api/indicadores/'
        + 'desarrolladores/jsonxml/INDICATOR/'
        + pop
        + '/es/0700/'
        + query
        + '/BISE/2.0/'
        + token
#        + '?type=json'
)
```

```python
# Se pasa como un request con metodo get
response = requests.get(url)
# Se le solicita el codigo de respuesta al servidor.    
status = response.status_code
```

```python
# Guarda la respuesta como una variable
content = response.json()
# Se filtra el json
data = content["Series"][0]["OBSERVATIONS"]
# Se accesa el diccionario con los datos
# pandas dataframe
df = pd.DataFrame(data)
```

```python
df
```

```python
response = requests.get(inegi)
data = json.loads(response.text)
data.keys()
df = pd.DataFrame(data['Series'])
df.explode(list('OBSERVATIONS'))
```

```python
catalog = pd.read_csv('https://www.inegi.org.mx/rnm/index.php/catalog/export/csv?ps=5000&collection[]=')
catalog['changed'] = pd.to_datetime(catalog['changed'])
catalog = catalog.set_index('id')
```

```python
catalog
```

```python
sep2023 = catalog.loc[catalog['changed'].between('2023-09-01', '2023-09-14')]
sep2023.head()
```

```python
url = 'https://www.inegi.org.mx/rnm/index.php/catalog/'

for id in sep2023.index:
    response = requests.get(url + str(id) + '/related-materials')
    print('Response code: ', response.status_code)
    soup = BeautifulSoup(response.text, 'html.parser')
    links = soup.find_all('a', class_ = 'download btn btn-outline-primary btn-sm btn-block')
    time.sleep(3)
```
