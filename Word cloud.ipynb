# Word cloud

In this notebook we will create a word cloud.

## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

## Create wordcloud instance

```python
# read csv file. Use specific cols and convert column to string
words = pd.read_csv('words.csv', usecols=['Creatividad']).Creatividad.str.cat()
```

```python
# create word cloud instance
wordcloud = WordCloud().generate(words)
# create chart
plt.imshow(wordcloud)
plt.axis("off")
```

## Customize word cloud

```python
wordcloud = WordCloud(background_color="#7393B3", max_words=100).generate(words)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

## Add a logo image to customized word cloud

```python
mask_img = np.array(Image.open("./meta.png"))

wc = WordCloud(background_color='#7393B3', mask=mask_img, contour_width=2, width=1000, height=800,
     contour_color='black', colormap='gnuplot').generate(words)
plt.axis('off')
plt.imshow(wc)
```
